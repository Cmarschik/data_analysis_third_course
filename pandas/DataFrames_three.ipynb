# Retrieve Rows by Index Label with .loc[]

```python
bond = pd.read_csv("jamesbond.csv", index_col = "Film")
bond.sort_index(inplace = True) # improves efficiency
bond.head(3)
```

```python
bond.loc["Goldfinger"] # used to extract rows since regular bracket syntax extracts coluns / returns a series for this single value
bond.loc["GoldenEye"]
#bond.loc["Sacred Bond"] # if incorrect you get an error
bond.loc["Casino Royale"] # returns all responses (whereas python dictionaries can't have duplicates) / df instead of series with multiple rows
```

```python
bond.loc["Diamonds Are Forever" : "Moonraker"] # extracts all between two movies / inclusiv values
bond.loc["GoldenEye":] # all moves with & beyond GoldenEye
bond.loc[:"On Her Majesty's Secret Service"] # all up unil this movie(inclusive)
```

```python
bond.loc[["Octopussy", "Moonraker"]] # use an embedded list to return multiple specified values / results in order requested in list
```

```python
bond.loc[["For Your Eyes Only", "Live and Let Die", "Gold Bond"]] # Gold bond is not real and returns null(NaN) values
```

# Retrive Rows by Index Position with .iloc[]

```python
bond = pd.read_csv("jamesbond.csv")
bond.head(3)
```

```python
bond.loc[15]
bond.iloc[15] # location and index position are the same in this example

bond.iloc[[15, 20]]
bond.iloc[:4]
bond.iloc[4:8] # 8 is exclusive
bond.iloc[20:]
```

# The .nsmallest() and .nlargest() Methods

```python
bond = pd.read_csv("jamesbond.csv", index_col = "Film")
bond.sort_index(inplace = True)
bond.head(3)
```

```python
bond.sort_values("Box Office", ascending = False).head()
```

```python
bond.nlargest(3, columns = "Box Office") # returns largest in a column / n = number of rows to extract / more efficient than sort_values

bond.nsmallest(2, columns = "Box Office")
```

```python
bond.nlargest(3, columns = "Budget")


bond.nsmallest(n = 6, columns = "Bond Actor Salary")
bond.nlargest(n = 3, columns = "Bond Actor Salary")
```

```python
bond["Box Office"].nlargest(5)

bond["Year"].nsmallest(2)
```

# Filtering with the where Method

```python
bond = pd.read_csv("jamesbond.csv", index_col = "Film")
bond.sort_index(inplace = True)
bond.head(3)
```

```python
mask = bond["Actor"] == "Sean Connery"
bond[mask]
```

```python
bond.where(mask) # accepts boolean series as argument / returns entire dataframe but only matching values are present...remaining are null(NaN)
```

```python
bond.where(bond["Box Office"] > 800)
```

```python
mask2 = bond["Box Office"] > 800
```

```python
bond.where(mask & mask2) # both conditions must be met
```

# The .query() Method

```python
# must have no spaces in column names to use the .query method
```

```python
bond = pd.read_csv("jamesbond.csv", index_col = "Film")
bond.sort_index(inplace = True)
bond.head(3)
```

```python
bond.columns = [column_name.replace(" ","_") for column_name in bond.columns] # replaces spaces with underscores in column names
bond.head()
```

```python
bond.query('Actor == "Sean Connery"')
bond.query("Director == 'Terence Young'")
bond.query('Actor != "Roger Moore"')
```

```python
bond.query("Box_Office > 600")
```

```python
bond.query("Actor == 'Roger Moore' or Director == 'John Glen'") # multiple conditions / use literal word 'and'/'or'
```

```python
bond.query("Actor not in ['Timothy Dalton', 'George Lazenby']") # for given actors (not) in bond movie
```

# A review of .apply() Method on Single Columns

```python
# applies something(function) to every value in a series
```

```python
bond = pd.read_csv("jamesbond.csv", index_col = "Film")
bond.sort_index(inplace = True)
bond.head(3)
```

```python
def convert_to_string_and_add_million(number):
    return str(number) + " MILLION!"
```

```python
columns = ["Box Office", "Budget", "Bond Actor Salary"]
 
    # loops through three columns doing the custom method to each value in each column and reassigning it
    
for col in columns:
    bond[col] = bond[col].apply(convert_to_string_and_add_million)
```
