# Outer Joins

```python
# combines values in either dataframe or both (or can combine those in either value but NOT both)
```

```python
week1 = pd.read_csv("Restaurant - Week 1 Sales.csv")
week2 = pd.read_csv("Restaurant - Week 2 Sales.csv")
customers = pd.read_csv("Restaurant - Customers.csv")
foods = pd.read_csv("Restaurant - Foods.csv")
```

```python
week1.head(2)
```

```python
week2.head(2)
```

```python
merged = week1.merge(week2, how = "outer", on = "Customer ID", suffixes = [" - Week 1", " - Week 2"], 
           indicator = True) # outer merge on a matching columns(full inclusive join by default)
# 'indicator' displays where the values are valid(where it's found / valid)
```

```python
merged["_merge"].value_counts() # returns how many of each value are occuring(right, left, or both)
```

```python
mask = merged["_merge"].isin(["left_only", "right_only"]) # returns True for left and right only's
```

```python
merged[mask] # returns dataframe of those not in both(exclusive outer join)
```

# Left Join

```python
# set 1 dataframe as the main focus / see if DF B is in DF A, if they don't exist, just put null, if they do, return A & what's in both
```

```python
# Right Join is the same (could just flip which DF is concidered left for the same result)
```

```python
week1 = pd.read_csv("Restaurant - Week 1 Sales.csv")
week2 = pd.read_csv("Restaurant - Week 2 Sales.csv")
customers = pd.read_csv("Restaurant - Customers.csv")
foods = pd.read_csv("Restaurant - Foods.csv")
```

```python
week1.head(3)
```

```python
foods.head(3)
```

```python
week1 = week1.merge(foods, how = "left", on = "Food ID", sort = True) # merges all food id info into the week1 (as long as they match a food id in week 1)
# sort parameter(as True) sorts the new DF on the given 'on' value
```

```python
week1.head(3)
```

# The left_on and right_on Parameters

```python
week1 = pd.read_csv("Restaurant - Week 1 Sales.csv")
week2 = pd.read_csv("Restaurant - Week 2 Sales.csv")
customers = pd.read_csv("Restaurant - Customers.csv")
foods = pd.read_csv("Restaurant - Foods.csv")
```

```python
week2.head(3)
```

```python
customers.head(3)
```

```python
week2 = week2.merge(customers, how = "left", left_on = "Customer ID", right_on = "ID", sort = True).drop("ID", axis = "columns") # mergest without matching column name
# left_on and right_on declare which two columns to merge on
# Call the .drop() method to remove one of the unneccesary matching columns we merged on
```

```python
week2.head(3)
```

# Merging by Indexes with the left_index and right_index Parameters 

```python
week1 = pd.read_csv("Restaurant - Week 1 Sales.csv")
week2 = pd.read_csv("Restaurant - Week 2 Sales.csv")
customers = pd.read_csv("Restaurant - Customers.csv", index_col = "ID")
foods = pd.read_csv("Restaurant - Foods.csv", index_col = "Food ID")
```

```python
customers.head(3)
```

```python
foods.head(3)
```

```python
sales = week1.merge(customers, how = "left", left_on = "Customer ID", right_index = True)
```

```python
sales.head(3)
```

```python
foods.head(3)
```

```python
sales = sales.merge(foods, how = "left", left_on = "Food ID", right_index = True)
```

```python
week1.head(3)
```

```python
week2.head(3)
```

```python
week1.merge(week2, how = "left", left_index = True, right_index = True, suffixes = [" - Week 1", " - Week 2"])
```
