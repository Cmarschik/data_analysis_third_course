# Bar Graphs

```python
google = data.DataReader(name = "GOOG", data_source = "yahoo", start = "2004-01-01", end = "2019-12-31") #import google data
google.head(3)
```

```python
def rank_performance(stock_price): #categorizes stock prices
    if stock_price <= 200:
        return "Poor"
    elif stock_price > 200 and stock_price < 500:
        return "Satisfactory"
    else:
        return "Stellar"
```

```python
google["Close"].apply(rank_performance).value_counts().plot(kind = "barh") # apply our function to all values in this series / return count of values
                                                        # ^returns plot / 'kind' parameter defines the type of graph
                                                        # "barh" makes bar graph horizontal
```

# Pie Charts

```python
apple = data.DataReader(name = "AAPL", data_source = "yahoo", start = "2012-01-01", end = "2019-12-31")
apple.head(3)
```

```python
apple["Close"].mean() # found overall mean for close
```

```python
def rank_performance(stock_price):
    if stock_price >= 118.77780614127028:
        return "Above Average"
    else:
        return "Below Average"
```

```python
plt.style.use("ggplot") #applies style "ggplot"
apple["Close"].apply(rank_performance).value_counts().plot(kind = "pie", legend = True) # plots info as a pie chart / with a legend parameter
```

# Histograms

```python
google = data.DataReader(name = "GOOG", data_source = "yahoo", start = "2004-01-01", end = "2019-12-31")
google.head(3)
```

```python
def custom_round(stock_price):
    return int(stock_price / 100.0) * 100 # creates integer that rounds to the closest 100
```

```python
google["High"].apply(custom_round).value_counts().sort_index() # returns stock price number in each value bucket
```

```python
google["High"].apply(custom_round).nunique() # returns how many buckets their are
```

```python
google["High"].apply(custom_round).plot(kind = "hist", bins = 13) # plots values as a histogram / bins sets buckets more accurate
```
